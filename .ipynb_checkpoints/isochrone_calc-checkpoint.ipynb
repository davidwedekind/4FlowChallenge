**Export to geojson**

```python
def to_geojson(geometry):
    
    points = str(geometry).replace('(','').replace(')','').replace('POLYGON ','').split(', ')
    
    coordinates = list()
    
    for string in points:
        
        lat_lon = string.split(' ')
        lat_lon[0] = float(lat_lon[0])
        lat_lon[1] = float(lat_lon[1])
        
        coordinates.append(lat_lon)
        
    geojson= list([coordinates])
    return {"type": "Polygon", "coordinates": geojson}
```

```python
feat = geojson.Feature(geometry=to_geojson(bike_delivery_area))
features = []
features.append(feat)
```

```python
feature_collection = geojson.FeatureCollection(features)
```

```python
feature_collection
```

```python
with open('delivery_area.geojson', 'w', encoding='utf-8') as f:
    json.dump(feature_collection, f, ensure_ascii=False)
```

## Read-In lor data

```python
polygons = df_lor.geometry.tolist()
```

## Share calculation

```python
polygons.append(bike_delivery_area)
```

**polygon fragmenting**

```python
boundaries = [LineString(list(shape(p).exterior.coords)) for p in polygons]
```

```python
union = unary_union(boundaries)
```

```python
result = [geom for geom in polygonize(union)]
```

```python
df_fragments = pd.DataFrame()
df_fragments['geom'] = pd.Series(result)
df_fragments.head(3)
```

```python
features = df_fragments.apply(lambda row: Feature(geometry=row.geom), axis=1).tolist()
feature_collection = FeatureCollection(features=features)
```

```python
with open('fragments.geojson', 'w', encoding='utf-8') as f:
    json.dump(feature_collection, f, ensure_ascii=False)
```

**determine intersections and intersection areas with lors**

```python
def get_intersect(x):
    
    intersect = x.intersection(bike_delivery_area)
    
    try:
        area = ops.transform(
            partial(
                pyproj.transform,
                pyproj.Proj(init='EPSG:4326'),
                pyproj.Proj(
                    proj='aea',
                    lat_1=intersect.bounds[1],
                    lat_2=intersect.bounds[3])),            
                intersect)
        
        return intersect, area.area
    
    except:
        
        return intersect, 0
        
df_lor[['ba_intersect_geom', 'ba_intersect_area']] = pd.DataFrame(df_lor.geometry.apply(lambda x: get_intersect(x)).tolist(), index=df_lor.index)
df_lor.head(2)
```

**calculate area**

```python
df_lor['intersect_share'] = round((df_lor.ba_intersect_area / df_lor.area)/((df_lor.ba_intersect_area / df_lor.area).max()),4)
```

```python
df_lor['pop_share'] = df_lor['pop'] / df_lor['pop'].sum()
```

```python
df_lor['share'] = df_lor['intersect_share'] * df_lor['pop_share']
```
